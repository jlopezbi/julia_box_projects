```julia
using PyPlot
```

```julia
function plot_diff(N, r, x_0a, x_0b)
    a = logisitic_loop(N, r, x_0a)
    b = logisitic_loop(N, r, x_0b)
    dif = abs(b-a)
    figure(1)
    plot(dif, "r.")
    xlabel("iterate n")
    ylabel("abs(A-B)")
    return dif
end
```

```julia
n = 500
r = 2.0
x_0a = 0.2
x_0b = 0.200001
dif_vec = plot_diff(n, r, x_0a, x_0b);
```
